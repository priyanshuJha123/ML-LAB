```python
import pandas as pd
import numpy as np
from sklearn import naive_bayes
print("Import cse21165..")
```

```python
data = pd.read_excel("student.xlsx")
print("Data loaded successfully...")
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample categorical data
sample = ['age','income','student','credi_rating','buys_computer']
# Initialize LabelEncoder
label_encoder = LabelEncoder()

for i in sample:
    data[i] = label_encoder.fit_transform(data[i])

print("Data encoded successfully...")

```

```python
data
```

```python
'''Prior probability = class_count/ len(data)'''
class_count = data['buys_computer'].value_counts()
data_length = len(data)
p_prob = class_count/data_length
print("Class count: ")
print(class_count)
print("Length of data frame: ")
print(data_length)
print("Prior probability : ",p_prob)
```

```python
'''A2.Calculate the class conditional densities for various features & classes. Observe if any class conditional density has zero values'''
from sklearn.naive_bayes import GaussianNB

X = data[['age','income','student','credi_rating']]
y = data['buys_computer'] 

# Initialize Gaussian Naive Bayes model
model = GaussianNB()

# Fit the model
model.fit(X, y)

# Class conditional densities for features and classes
class_conditional_densities = {}

# Calculate class conditional densities for each feature and class
for class_label in set(y):
    class_conditional_densities[class_label] = {}
    for feature_idx in range(X.shape[1]):
        feature_values = X[y == class_label, feature_idx]
        class_conditional_densities[class_label][feature_idx] = {
            'mean': feature_values.mean(),
            'std': feature_values.std()
        }

# Display class conditional densities
for class_label, feature_densities in class_conditional_densities.items():
    print(f"Class {class_label} conditional densities:")
    for feature_idx, density_info in feature_densities.items():
        print(f"Feature {feature_idx}: Mean = {density_info['mean']}, Std = {density_info['std']}")
    print()

# Check for zero values in class conditional densities
zero_densities = False
for class_label, feature_densities in class_conditional_densities.items():
    for feature_idx, density_info in feature_densities.items():
        if density_info['std'] == 0:
            print(f"Class {class_label}, Feature {feature_idx}: Zero density found!")
            zero_densities = True

if not zero_densities:
    print("No class conditional density has zero values.")
```

```python
'''A4. Build a Na√Øve-Bayes (NB) classifier for the above given data.'''
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score,confusion_matrix
from sklearn.naive_bayes import GaussianNB

# Creating the instance of GaussianNB
model = GaussianNB()

X = data[['age','income','student','credi_rating']]
y = data['buys_computer']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.6)

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
confusion = confusion_matrix(y_test,y_pred)
```

```python
print("Accuracy of the model : ",accuracy)
print("Precision of the model: ",precision)
print("Recall of the model   : ",recall)
print("Confusion Matrix      : \n",confusion)
```
